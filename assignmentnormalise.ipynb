```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from IPython.display import display, HTML
```

```python

# Load file
df = pd.read_csv('data.csv', index_col=0)
# Reset the index and change it to start from 1
df = df.reset_index(drop=True)
df.index = np.arange(1, len(df) + 1)

df.sample(n=60, random_state=1)
```

```python
df = df[['Title', 'Company', 'Location',  'Classification', 'SubClassification', 'LowestSalary', 'HighestSalary', 'Date']]
df.head(60)
```

```python
# Filter rows where LowestSalary is greater than 100
df = df[df['LowestSalary'] > 50]
df.head(60)
```

```python
# Filter rows where Location is either 'Sydney' or 'Perth'
df = df[(df['Location'] == 'Sydney') | (df['Location'] == 'Perth')]
# display 50/50 of each location
df_sydney = df[df['Location'] == 'Sydney']
df_perth = df[df['Location'] == 'Perth']
df_sydney = df_sydney.sample(n=30, random_state=1)
df_perth = df_perth.sample(n=30, random_state=1)
df_combined = pd.concat([df_sydney, df_perth])
df_combined.head(60)
```

```python
df_filtered = df_combined.reset_index(drop=True)
df_filtered.index = np.arange(1, len(df_filtered) + 1)
df_filtered.head(60)
```

```python
# Normalize Date
df_filtered['Date'] = pd.to_datetime(df_filtered['Date']).dt.date
df_filtered.head(60)
```

```python
df_filtered = df_filtered.sort_values(by=['Location', 'Classification', 'LowestSalary'], ascending=[True, True, True])
df_filtered = df_filtered.reset_index(drop=True)
df_filtered.index = np.arange(1, len(df_filtered) + 1)
df_filtered.head(60)
```

```python
def display_with_style(title):
    display(HTML(f"<h2 style='font-size:20px;'>{title}</h2>"))


display_with_style("Description: ")
df_Description = df_filtered[['Title', 'Company', 'Location',  'Date']]
df_Description.head(60)
```

```python
display_with_style("Salary:")
df_Salary = df_filtered[['Location', 'Classification', 'SubClassification', 'LowestSalary', 'HighestSalary', 'Date']]
df_Salary.head(60)
```

```python
# range of Salary
df_Salary['SalaryRange'] = df_Salary['HighestSalary'] - df_Salary['LowestSalary']
df_Salary = df_Salary[['Location', 'Classification', 'SubClassification', 'LowestSalary', 'HighestSalary', 'SalaryRange', 'Date']]
#spliting the table in half
df_Salary.tail(30)
```

```python
# Calculate the IQR for SalaryRange
Q1 = df_Salary['SalaryRange'].quantile(0.25)
Q3 = df_Salary['SalaryRange'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_Outliners = df_Salary[(df_Salary['SalaryRange'] >= lower_bound) & (df_Salary['SalaryRange'] <= upper_bound)]
df_Outliners = df_Outliners.reset_index(drop=True)
df_Outliners.index = np.arange(1, len(df_Outliners) + 1)
df_Outliners.head(60)
```

```python
df_PerthS = df_Outliners[df_Outliners['Location'] == 'Perth']
df_PerthS.head(30)
```

```python
df_SydneyS = df_Outliners[df_Outliners['Location'] == 'Sydney']
df_SydneyS.head(30)
```
