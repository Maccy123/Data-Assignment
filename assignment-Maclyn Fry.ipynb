# 1.2.Normalize and clean the data.

### Calculate the avarage salary using df.assign (1 mark)

```python
df = df.assign(SalaryMedian=lambda x: x.LowestSalary + (x.HighestSalary - x.LowestSalary) / 2 )
df.head()
```

```python
Mean = df["SalaryMedian"].mean()
print("Salary Mean: ", Mean)
```

### Clean Id column using regex (1 mark)

```python
##Hint replace(to_replace=r'&.*', value='', regex=True)

df["Id"] = df["Id"].replace(to_replace=r'&.*', value='', regex=True)
df.head()
```

### Clearn date using regex (1 mark)

```python
#Your code to clean date
df["Date"] = df["Date"].replace(to_replace=r'T.*', value='', regex=True)
df.head()
```

### Change type of related columns (1 mark)

```python
# Replace non-numeric values with NaN
df['Id'] = pd.to_numeric(df['Id'], errors='coerce')

# Drop rows with NaN values in 'Id' column
df = df.dropna(subset=['Id'])

# Convert 'Id' to int64
df['Id'] = df['Id'].astype(np.int64)

# Set column "Date" to type datetime64[ns]
df["Date"] = pd.to_datetime(df['Date'])
# Data types after data cleaning and conversions
print(df.dtypes)
```

```python
#Save the cleaned and preprocessed dataset
df.to_csv("preprocessed_data.csv")
```

# 1.3. Job metadata

```python
colors = ['#1F77B4', '#AEC7E8', '#FF7F0E', '#FFBB78', '#2CA02C', '#98DF8A', '#B5CF6B', '#FF9896','#9467BD', '#C5B0D5', '#BCBD22']
```

```python
#Group by average salary and count the numbers
averageSalaryDf = df.groupby('SalaryMedian').count()
averageSalaryDf['SalaryMedian'] = (averageSalaryDf.index*1000).astype(float).map('{:,.0f}'.format)
averageSalaryDf['Count'] = averageSalaryDf["Id"]
averageSalaryDf
```

### Bar chart to display the correlations between “AverageSalary” and the total jobs of each range (2 marks)

```python
ax = averageSalaryDf[['SalaryMedian', 'Count']].plot.bar(rot=0)
```

### Pie chart to display the relationship between job types and the number of jobs (2 marks).

```python
JobNumDf = df.groupby('JobType').count()
JobNumDf['Count'] = JobNumDf["Id"]
JobNumDf = JobNumDf[['Count']]
plot = JobNumDf.plot.pie(title = 'Job distribution by job type' ,y = 'Count' , figsize=(4, 4), autopct='%1.1f%%', legend = False, colors=colors)
```

# Study market by locations

```python
locationDf = df[df['Location'].isna()==False]
locationDf = locationDf.groupby('Location').count()
locationDf['Location'] = locationDf.index
locationDf['Count'] = locationDf["Id"]
locationDf = locationDf[locationDf['Count'] > 400]
locationDf
```

### Get top 20 big cities and use bar chart to display market size in each city (2 marks).

```python
locationDf = locationDf.sort_values(by=['Count'], ascending= True)
ax1 = locationDf.plot.barh(title = 'Market size of each cities', x='Location', y='Count', figsize=(15, 30), legend = False)
ax1.grid(True, linestyle='--', alpha=0.7)
```

### Visualize the market share between the cities (2 marks)

```python
#Get top 10 big cities
locationDf = locationDf.sort_values(by=["Count"], ascending =False)
locationTop10Df = locationDf[:10].copy()
#others
new_row = pd.DataFrame(data = {
    'Location' : ['Others'],
    'Count' : [locationDf["Count"][10:].sum()]
})
locationTop10Df = pd.concat([locationTop10Df, new_row],sort=False)
locationTop10Df.index = locationTop10Df['Location']
```

```python

MarketDf = locationTop10Df[['Count']]
plot = MarketDf.plot.pie(title = 'Market share of sectors', y = 'Count' , figsize=(4, 4), autopct='%1.1f%%', legend=False, colors=colors )
```

### Visualize the salary distribution in top biggest cities (2 marks)

```python
# Assuming df is already defined and contains the necessary data

# Filter the dataframe for the three cities
BiggestCitydf = df[df['Location'].isin(['Sydney', 'Brisbane', 'Melbourne'])]

# Group by 'SalaryMedian' and 'Location', and count the occurrences
BigCityDf = BiggestCitydf.groupby(['SalaryMedian', 'Location']).count()

# Extract the 'Count' column and reset the index to make it easier to work with
BigCityDf['Count'] = BigCityDf['Id']
BigCityDf = BigCityDf['Count'].reset_index()

# Create separate DataFrames for each city
Bris = BigCityDf[BigCityDf['Location'] == 'Brisbane'][['SalaryMedian', 'Count']]
Syd = BigCityDf[BigCityDf['Location'] == 'Sydney'][['SalaryMedian', 'Count']]
Melb = BigCityDf[BigCityDf['Location'] == 'Melbourne'][['SalaryMedian', 'Count']]

# Merge the DataFrames on 'SalaryMedian'
merged_df = Bris.merge(Syd, on='SalaryMedian', how='outer', suffixes=('_Bris', '_Syd'))
merged_df = merged_df.merge(Melb, on='SalaryMedian', how='outer')
merged_df.rename(columns={'Count': 'Count_Melb'}, inplace=True)

# Fill NaN values with 0
merged_df.fillna(0, inplace=True)

# Set 'SalaryMedian' as the index
merged_df.set_index('SalaryMedian', inplace=True)

# Plotting the result
ax = merged_df.plot.bar(stacked=True)
ax.set_xlabel("Salary Median")
ax.set_ylabel("Count")
ax.set_title("Salary Distribution in Top Biggest Cities")
```

# 1.5 Study by sectors

### The number of postings for each sector. (2 marks)

```python
ClassDf = df[df['Classification'].isna()==False]
ClassDf = ClassDf.groupby('Classification').count()
ClassDf['Classification'] = ClassDf.index
ClassDf['Count'] = ClassDf["Id"]
ClassDf = ClassDf[ClassDf['Count'] > 400]
ClassDf = ClassDf.sort_values(by=['Count'], ascending= True)
ax1 = ClassDf.plot.barh(title = 'Market size of each cities', x='Classification', y='Count', figsize=(15, 30), legend = False)
ax1.grid(True, linestyle='--', alpha=0.7)
```

### Find the main spotlight sub-sectors in each sector (1 mark)

```python
# Pie char for job market share of classifications
from matplotlib import cm

df_pie = ClassDf['Classification'].groupby(ClassDf['Classification'])

df_pie = df_pie.count().sort_values(ascending=False).to_frame()
df_pie['Value'] = df_pie["Classification"]
df_pie['Classification'] = df_pie.index
df2 = df_pie[:10].copy()
new_row = pd.DataFrame(data = {
    'Classification' : ['Others'],
    'Value' : [df_pie["Value"][10:].sum()]
})

df2 = pd.concat([df2, new_row])
df2.index = df2['Classification']

plot = df2.plot.pie(title = 'Market share of sectors', y = 'Value' , figsize=(6, 6), legend=False, colors=colors )
```

### Visualize the salary range in the sectors/sub-sectors (1 mark)

```python
SubClassDf = df[df['Classification'].isna()==False]
SubClassDf = SubClassDf.groupby(['Classification', 'SubClassification']).count()
SubClassDf['Classification', 'SubClassification'] = SubClassDf.index
SubClassDf['Count'] = SubClassDf["Id"]
SubClassDf = SubClassDf[SubClassDf['Count'] > 400]
SubClassDf = SubClassDf.sort_values(by=['Count'], ascending= True)
ax1 = SubClassDf.plot.barh(title = 'Market size of each cities', y='Count', figsize=(15, 30), legend = False)
ax1.grid(True, linestyle='--', alpha=0.7)

```
